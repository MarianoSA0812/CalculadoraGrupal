```python
def mostrar_menu(nombre_alumno):
    print("Bienvenido al menú principal,", nombre_alumno)
    print("Elija una de las siguientes opciones:")
    print("1. ASISTENCIA")
    print("2. NOTAS")
    print("3. Salir")

def tomar_asistencia(nombre_alumno):
    print("Has seleccionado la Opción 1: Asistencia")
    print("Ingrese una opción:")
    print("1. Presente")
    print("2. Ausente")

    opcion_asistencia = input()

    if opcion_asistencia == "1":
        print("El alumno", nombre_alumno, "está Presente")
    elif opcion_asistencia == "2":
        print("El alumno", nombre_alumno, "está Ausente")
    else:
        print("Opción inválida")

def tomar_notas(nombre_alumno):
    print("Has seleccionado la Opción 2: Notas")
    print("Ingrese las notas (1 al 10). Ingrese '0' para finalizar.")

    total_notas = 0
    suma_notas = 0

    while True:
        nota_str = input("Ingrese una nota: ")

        if nota_str == "0":
            break

        nota = int(nota_str)

        if 1 <= nota <= 10:
            suma_notas += nota
            total_notas += 1
        else:
            print("Nota inválida. Ingrese un número del 1 al 10.")

    promedio = suma_notas / total_notas if total_notas > 0 else 0

    if promedio <= 4:
        print("El estado del alumno", nombre_alumno, "es Libre")
        print("El promedio del alumno", nombre_alumno, "es de:", format(promedio, ".2f"))
    elif promedio >= 7:
        print("El estado del alumno", nombre_alumno, "es Promocionado")
        print("El promedio del alumno", nombre_alumno, "es de:", format(promedio, ".2f"))
    else:
        print("El estado del alumno", nombre_alumno, "es Regular")
        print("El promedio del alumno", nombre_alumno, "es de:", format(promedio, ".2f"))

def main():
    nombre_alumno = input("Ingrese el nombre del alumno: ")

    opcion = 0

    while opcion != 3:
        mostrar_menu(nombre_alumno)

        opcion = int(input())

        if opcion == 1:
            tomar_asistencia(nombre_alumno)
        elif opcion == 2:
            tomar_notas(nombre_alumno)
        elif opcion == 3:
            print("Has seleccionado la Opción 3: Salir")
            break
        else:
            print("Opción inválida")

        input("Presiona Enter para continuar...")
        print()

if __name__ == "__main__":
    main()
```
